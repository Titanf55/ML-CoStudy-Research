# Handling null values

### there are 272 null values in wind gust that needs to be filled

```python
Partially_cloudy_mean = df[df['conditions'] == 'Partially cloudy']['windgust'].mean()
rain_mean = df[df['conditions'] == 'Rain']['windgust'].mean()
clear_mean = df[df['conditions'] == 'Clear']['windgust'].mean()
```

```python
print("the mean of windgust for 'Partially cloudy'= ",Partially_cloudy_mean)
print("the mean of windgust for 'rain'= ",rain_mean)
print("the mean of windgust for 'Clear'= ",clear_mean)
```

```python
df.loc[df['conditions'] == 'Partially cloudy', 'windgust'] = df.loc[df['conditions'] == 'Partially cloudy', 'windgust'].fillna(Partially_cloudy_mean)
df.loc[df['conditions'] == 'Rain', 'windgust'] = df.loc[df['conditions'] == 'Rain', 'windgust'].fillna(rain_mean)
df.loc[df['conditions'] == 'Clear', 'windgust'] = df.loc[df['conditions'] == 'Clear', 'windgust'].fillna(clear_mean)
```

# Train Test Split

```python
# Splttting data in train and test 

train = df.loc["2014-01-01":"2021-12-31"]
test = df.loc["2022-01-01":"2023-12-31"]
```

```python
# Input column
input_cols = list(train.columns)[0:-1]
target_col = 'conditions'
```

```python
print(input_cols)
print(target_col)
```

```python
# making a copy of Traing data

train_inputs = train[input_cols].copy()
train_targets = train[target_col].copy()
```

```python
# making a copy of Test data

test_inputs = test[input_cols].copy()
test_targets = test[target_col].copy()
```

```python
# Differentiating numerical columns and categorical column

numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes('object').columns.tolist()
```

```python
print("numeric columns are ",numeric_cols)
```

```python
print("numeric categorical are ",categorical_cols)
```

# Standardizing values of Input columns

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df[numeric_cols])
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].describe()
```

# Encoding categorical data in Input columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
encoder.fit(df[categorical_cols])
```

```python
encoder.categories_
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
print(encoded_cols)
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

# Encoding categorical data in Output columns

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[["Clear","Partially cloudy","Rain"]],dtype=int)
```

```python
oe.fit(df[["conditions"]])
```

```python
train_data_reshaped = np.array(train_targets).reshape(-1, 1)
test_data_reshaped = np.array(test_targets).reshape(-1, 1)
```

```python
train_targets = oe.transform(train_data_reshaped)
test_targets = oe.transform(test_data_reshaped)
train_targets = train_targets.reshape(-1)
test_targets = test_targets.reshape(-1)
```

```python
train_targets.shape
```

```python
test_targets.shape
```

```python
X_train=train_inputs[numeric_cols+encoded_cols]
X_test=test_inputs[numeric_cols+encoded_cols]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Applying boosting algorithms without smot

## 1. Gradient boosting

```python
model = GradientBoostingClassifier(
    random_state=40,
    max_depth=8,             # Maximum depth of the individual estimators
    n_estimators=300,        # Number of boosting stages
    learning_rate=0.2,       # Learning rate
    criterion="squared_error" # Criterion for splitting at each node
)
```

```python
model.fit(X_train,train_targets)
```

```python
X_train_pred = model.predict(X_train)
```

```python
print(accuracy_score(X_train_pred,train_targets))
print(confusion_matrix(X_train_pred,train_targets))
print(classification_report(X_train_pred,train_targets))
```

```python
X_test_pred = model.predict(X_test)
```

```python
print(accuracy_score(X_test_pred,test_targets))
print(confusion_matrix(X_test_pred,test_targets))
print(classification_report(X_test_pred,test_targets))
```

## 2. Catboost

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(
    iterations=300,          # Number of boosting iterations
    depth=8,                 # Depth of the tree
    learning_rate=0.2,       # Learning rate
    random_seed=40,          # Random seed for reproducibility
    loss_function='MultiClass',  # For multiclass classification, # Loss function (for multiple classification)
    eval_metric='Accuracy',  # Evaluation metric
    verbose=100              # Print updates every 100 iterations
)
```

```python
model.fit(X_train,train_targets)
```

```python
X_train_pred = model.predict(X_train)
```

```python
print(accuracy_score(X_train_pred,train_targets))
print(confusion_matrix(X_train_pred,train_targets))
print(classification_report(X_train_pred,train_targets))
```

```python
X_test_pred = model.predict(X_test)
```

```python
print(accuracy_score(X_test_pred,test_targets))
print(confusion_matrix(X_test_pred,test_targets))
print(classification_report(X_test_pred,test_targets))
```
